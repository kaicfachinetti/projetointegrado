## Etiquetando labels

```python
def etiquetar_labels(df):
  '''
  Essa função irá etiquetar as labels do conjunto de dados.
  '''
  df.loc[df['Label'] == 1, 'Label'] = 1
  df.loc[df['Label'] == -1, 'Label'] = 0

  return df
```

```python
df = etiquetar_labels(df)
```

## Separando conjunto treino e teste

```python
def separa_treino_teste(df):
  '''
  Essa função irá separar o conjunto de dados em treino e teste.
  '''
  #Definindo tamanho dos dados de treino e dados de teste:
  train_size = int(len(df) * 0.7)
  val_size = len(df) - train_size

  #Separando dados de treino e dados de teste:
  train = df.iloc[:train_size,:]
  val = df.iloc[train_size:,:]

  return train, val
```

## Preparando DataSet

```python
def preparar_dataset(df):
  '''
  Essa função irá transformar o dataset inserido em um array de X com os valores de fechamento dos últimos 15 dias e um array de y (labels)
  '''
  X = []
  y = []

  for i in range(len(df)):

    X.append([]) #Cria a lista que receberá, ordenado por data, os valores dos preços das ações.

    for j in range(16,1,-1): #Adiciona à lista X, ordenado do mais antigo ao mais novo, os valores dos preços das ações.
      X[i].append(df.iloc[i,j])

    y.append(df.iloc[i, 1]) #Adiciona a Label à lista y

  return np.array(X), np.array(y)
```

## Contruindo classe LSTMClassifier

```python
class LSTMClassifier():
  def __init__(self, n_hidden_layers, n_neurons, dropout_rate):
    self.n_hidden_layers = n_hidden_layers
    self.n_neurons = n_neurons
    self.dropout_rate = dropout_rate
    self.model = self.build_model()

  def build_model(self):
    '''
    Essa função irá contruir o modelo LSTM.
    '''
    self.model = Sequential()

    if self.n_hidden_layers <= 1:
      #Adicionando camada LSTM:
      self.model.add(LSTM(units=self.n_neurons, input_shape=(16,1), return_sequences=False, kernel_initializer=tf.keras.initializers.GlorotUniform()))
      self.model.add(Dropout(self.dropout_rate))

      #Adicionando camada de saída:
      self.model.add(Dense(1, activation='sigmoid'))

      return self.model

    else:
      for i in range(self.n_hidden_layers-1):
        if i == 0:
          #Adiciona camada LSTM:
          self.model.add(LSTM(units=self.n_neurons, return_sequences=True, input_shape=(16,1), kernel_initializer=tf.keras.initializers.GlorotUniform()))
          self.model.add(Dropout(self.dropout_rate))

        else:
          self.model.add(LSTM(units=self.n_neurons, return_sequences=True, kernel_initializer=tf.keras.initializers.GlorotUniform()))
          self.model.add(Dropout(self.dropout_rate))

      #Adiciona última camada oculta de LSTM:
      self.model.add(LSTM(units=self.n_neurons, return_sequences=False, kernel_initializer=tf.keras.initializers.GlorotUniform()))
      self.model.add(Dropout(self.dropout_rate))

      #Adiciona camada de saída:
      self.model.add(Dense(1, activation='sigmoid'))

      return self.model

  def compile(self, optimizer, loss, metrics):
    '''
    Essa função irá compilar o modelo.
    '''

    self.model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

    return self.model

  def fit(self, X, y, epochs, batch_size, validation_data, callbacks):
    '''
    Essa função irá treinar o modelo LSTM.
    '''

    #Treinando o modelo:
    self.history = self.model.fit(X, y, epochs=epochs, batch_size=batch_size, validation_data=validation_data, callbacks=callbacks)

    return self.model

  def predict(self, X, y_true):
    '''
    Essa função irá realizar a predição do modelo.
    '''
    # Realiza a predição
    y_pred = self.model.predict(X)

    #Threshold:
    #Em alguns treinamentos, as probabilidades previstas para todas as amostras fica menor do que 0.5 e por isso calcularemos um threshold ideal que pode ser diferente de 0.5
    from sklearn.metrics import roc_curve
    fpr, tpr, thresholds = roc_curve(y_true, y_pred)
    optimal_idx = np.argmax(tpr - fpr)
    optimal_threshold = thresholds[optimal_idx]

    #Binariza:
    for i in range(len(y_pred)):
      if y_pred[i] > optimal_threshold:
        y_pred[i] = 1
      else:
        y_pred[i] = 0

    return y_pred
```

### Treinando LSTMClassifier

```python
lstm_classifier = LSTMClassifier(n_hidden_layers=3, n_neurons=16, dropout_rate=0.3)
```

```python
lstm_classifier.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
checkpoint = tf.keras.callbacks.ModelCheckpoint('vale3_model.keras', monitor='val_accuracy', mode='auto', verbose=1, save_best_only=True)
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=15, verbose=1, mode='auto')
lstm_classifier.fit(X_train, y_train, epochs=50, batch_size=8, validation_data=(X_val, y_val), callbacks=[checkpoint, early_stopping])
```

```python
plot_model(lstm_classifier.model, show_shapes=True, show_layer_names=True)
```

```python
#Avaliação dos pesos das camadas ocultas:
print("Pesos e Biases das Camadas:")
for i, layer in enumerate(lstm_classifier.model.layers):
    print(f"\nCamada {i + 1} - {layer.name}")
    weights = layer.get_weights()
    if weights:  # Se a camada possui pesos treináveis
        for j, weight in enumerate(weights):
            print(f"Peso {j + 1}: Forma {weight.shape}")
            print(f"Valores: {weight}")  # Exibe os valores diretamente
    else:
        print("Esta camada não possui pesos treináveis.")
```
